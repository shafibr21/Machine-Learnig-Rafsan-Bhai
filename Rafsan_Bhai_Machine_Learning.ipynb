```python
from sklearn.preprocessing import LabelEncoder

encoder=LabelEncoder()
dataset["WindGustDir"]=encoder.fit_transform(dataset["WindGustDir"])
dataset["WindDir9am"]=encoder.fit_transform(dataset["WindDir9am"])
dataset["WindDir3pm"]=encoder.fit_transform(dataset["WindDir3pm"])
dataset["RainToday"]=encoder.fit_transform(dataset["RainToday"])
dataset["RainTomorrow"]=encoder.fit_transform(dataset["RainTomorrow"])
```

#Data Spliting

```python
dataset.iloc[1:4,3:5]    #[rowstart,rwoend-1,coloumstart:coloumend-1]
```

```python
X=dataset.iloc[:, 11:21] #features

Y=dataset["RainTomorrow"] #Target
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2)
```

#Training Phase

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=0)
rf.fit(X_train,Y_train)
y_pred = rf.predict(X_test)             #model predicted
```

# Accuracy 

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(Y_test,y_pred))
```
